Напиши функцию, которая принимает две матрицы — V и C (матрица вакансий и кандидатов соответственно). Они будут размера [num_vacancies, 5] и [num_candidates, 5]

Каждое значение в матрице — (i, j) — насколько кандидат/вакансия i соответствует характеристике j.  Этот вес от 0 до 1, мы его получим из берта косинусным сходством

Тебе нужно сделать следующее:
Для каждой пары (кандидат, вакансия) вернуть скор их схожести. Для кандидата c и вакансии v считается следующим образом:
\sum_{i=1}^{5} {w_i^c * w_i^v + (1- w_i^c) * (1-w_i^v)}

То есть должн вернуть матрицу размера [num_vacancies., num_candidates], где каждое значение (i, j) — скор схожести вакансии i с кандидатом j

скорее всего это что-то типа формулы V.dot(C.T) + (1-V).dot(1 - C.T)

```python
import numpy as np

# Function to calculate cosine similarity between two vectors
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    return dot_product / (norm1 * norm2)

# Function to calculate the similarity score between a candidate and a vacancy
def calculate_similarity_score(candidate, vacancy):
    score = 0
    for i in range(5):
        w_c = candidate[i]
        w_v = vacancy[i]
        score += w_c * w_v + (1 - w_c) * (1 - w_v)
    return score

# Generate random data for the example
num_vacancies = 10
num_candidates = 8

vacancies_matrix = np.random.rand(num_vacancies, 5)
candidates_matrix = np.random.rand(num_candidates, 5)

# Create similarity matrix (vacancy x candidate)
similarity_matrix = np.zeros((num_vacancies, num_candidates))

# Calculate similarity for each vacancy-candidate pair
for i in range(num_vacancies):
    for j in range(num_candidates):
        similarity_matrix[i, j] = calculate_similarity_score(
            candidates_matrix[j], vacancies_matrix[i]
        )

# Print the similarity matrix (vacancy x candidate)
print("Similarity matrix (vacancy x candidate):")
print(similarity_matrix)
```
